````python
import PyPDF2
from langchain_google_genai import GoogleGenerativeAI
from getpass import getpass
import os
from langchain.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough
````

````python
filename = "./Prompt Engineering.pdf"
````

````python
with open(filename, "rb") as f:
    reader = PyPDF2.PdfReader(f)
    
    study_material = ""
    for page in reader.pages:
        study_material += page.extract_text()

print(study_material)
````

````python
gemini_key = getpass("Enter your key: ")
````

````python
os.environ["GOOGLE_API_KEY"] = gemini_key
````

````python
# Updated parameter name from model_name to model:
gemini = GoogleGenerativeAI(model='gemini-2.5-flash-lite', temperature=0)
````

````python
summarize_prompt_template = """ Act as a teacher. Your job is to review the provided content delimited in backticks,
                                and generate a summary from it.
                        Content: ```{study_material}```"""
````

````python
summarize_prompt = ChatPromptTemplate.from_template(summarize_prompt_template)
````

````python
summary_chain = (
    summarize_prompt
    |
    gemini
    |
    StrOutputParser()
)
````

````python
summary_chain.invoke({"study_material": study_material})
````

````python
qna_generator_prompt_template = """ Act as a moderator. Your task is to generate a set of multiple choice question and answer,
                                    from the provided summary delimited in backticks.
                                    Example:
                                        What is the primary goal of prompt engineering in agent-based systems?
                                            a) To optimize agent memory
                                            b) To refine inputs for better output control
                                            c) To improve agent hardware
                                            d) To increase computational power
                                            Answer: b) To refine inputs for better output control
                                    Summary:
                                    ```{summary}```
                                """
````

````python
qna_generator_prompt = ChatPromptTemplate.from_template(qna_generator_prompt_template)
````

````python
qna_generator_chain = (
    qna_generator_prompt
    |
    gemini
    |
    StrOutputParser()
)
````

````python
full_chain = (
    RunnablePassthrough.assign(summary=summary_chain)
    |
    RunnablePassthrough.assign(quiz=qna_generator_chain)
)
````

````python
result = full_chain.invoke({"study_material": study_material})
````
