```python
current_price = company_info.get("currentPrice")
previous_close = company_info.get("previousClose")
if current_price is None or previous_close is None:
    price_change = None
    percentage_change = None
else:
    price_change = previous_close - current_price
    percentage_change = price_change/previous_close*100
{
    "market_state": company_info.get("marketState", "N/A"),
    "current_price": current_price,
    "previos_close": previous_close,
    "price_change": price_change,
    "percentage_change": percentage_change,
    "currency": company_info.get("currency", "N/A"),
    "day_high": company_info.get("dayHigh", "N/A"),
    "day_low": company_info.get("dayLow", "N/A"),
    "volume": company_info.get("volume", "N/A"),
    "exchange": company_info.get("exchange", "N/A")
}
```
